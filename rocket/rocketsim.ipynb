Rocket Start without Gravity

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def v_rocket_no_gravity(t):
    fe = 2.0
    rb = 1.0
    mn = 10.0
    ms = 100.0
    v = (-fe) * (1.0/rb * np.log(np.abs(rb*t - mn -ms)))
    return v    
```

```python
x = np.linspace(0, 100, 100)
y = v_rocket_no_gravity(x)

plt.plot(x, y)
plt.title('Plot of V no Gravity')
plt.xlabel('t')
plt.ylabel('v')
plt.grid(True)
plt.show()
```
