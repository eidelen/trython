# Lamp on a wheel

```python
import numpy as np
import math
import matplotlib.pyplot as plt

alphas = np.linspace(0.0, 6 * math.pi, 100)
R = 1.0

xs = []
ys = []
for alpha in alphas:
    xs.append(R * alpha + R * math.sin(alpha))
    ys.append(R + R * math.cos(alpha))
    
plt.figure()
plt.plot(xs, ys)
plt.axis('equal')
plt.show()

```
