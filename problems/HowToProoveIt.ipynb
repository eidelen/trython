```python
import pandas as pd
from sympy import isprime
import math
```

```python
# 2) prime conjecture 3^n - 1 and 3^n-2^n

ns = []
p = []
c1 = []
p1 = []
c2 = []
p2 = []

for n in range(2, 100):
    con1 = 3**n -1
    con2 = 3**n - 2**n
    ns.append(n)
    p.append('Y' if isprime(n) else 'N')
    c1.append(con1)
    p1.append('Y' if isprime(con1) else 'N')
    c2.append(con2)
    p2.append('Y' if isprime(con2) else 'N')

df = pd.DataFrame({
    "n": ns,
    "p": p,
    "3^n-1": c1,
    "p1": p1,
    "3^n-2^n": c2,
    "p2": p2
})

df

```

```python
# ex 4
n = 5
x = math.factorial(n+1) + 2
for q in range(x, x+n):
    print(q, ': Prime' if isprime(q) else ' NoPrime')
```

```python
# ex 4, longest sequence of non prime
nonprime_seq = []
for z in range(1, 100):
    if not isprime(z):
        nonprime_seq.append(z)
    else:
        print(len(nonprime_seq), nonprime_seq)
        nonprime_seq = []

```

```python
# ex 5, perfect numbers
def check_perfect(n):
    s = 0
    for l in range(1, n):
        if n%l == 0:
            s += l
    return s == n

for c in range(1, 30):
    print(c, check_perfect(c))

n1 = 5
p1 = 2**(n1-1) * (2**n1 - 1)
print(p1, check_perfect(p1))

n2 = 7
p2 = 2**(n2-1) * (2**n2 - 1)
print(p2, check_perfect(p2))
```

```python
#ex 7
from sympy import divisors
def sum_divisors(n):
    divs = divisors(n)
    divs.pop()
    return sum(divs)
print(sum_divisors(220))
print(sum_divisors(284))
```
