```python
import pandas as pd
from sympy import isprime
```

```python
# 2) prime conjecture 3^n - 1 and 3^n-2^n

ns = []
p = []
c1 = []
p1 = []
c2 = []
p2 = []

for n in range(2, 20):
    con1 = 3**n -1
    con2 = 3**n - 2**n
    ns.append(n)
    p.append('Y' if isprime(n) else 'N')
    c1.append(con1)
    p1.append('Y' if isprime(con1) else 'N')
    c2.append(con2)
    p2.append('Y' if isprime(con2) else 'N')

df = pd.DataFrame({
    "n": ns,
    "p": p,
    "3^n-1": c1,
    "p1": p1,
    "3^n-2^n": c2,
    "p2": p2
})

df

```
